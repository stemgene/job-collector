```python
html_str = """
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <div>
        <ul>
            <li class="item-1">
                <a href="link1.html">First</a>
            </li>
            <li class="item-2">
                <a href="link2.html">Second</a>
            </li>
            <li class="item-3">
                <a href="link3.html">Third</a>
            </li>
            <li class="item-4">
                <a href="link4.html">Fouth</a>
            </li>
            <li class="item-5">
                <a href="link5.html">Fifth</a>
            </li>
        </ul>
    </div>
</body>
</html>
"""
```

```python
# 转换数据类型
data = parsel.Selector(html_str)
data
```

```python
# 从根节点开始获取所有<a>标签
result1 = data.xpath("/html/body/div/ul/li/a").extract() # ['<a href="link1.html">First</a>', '<a href="link1.html">Second</a>', '<a href="link1.html">Third</a>', '<a href="link1.html">Fouth</a>', '<a href="link1.html">Fifth</a>'] 
# 跨节点，获取所有<a>标签
result2 = data.xpath("//a").extract()
# 选取当前节点。使用场景需要对选取的标签的下一级标签进行多次提取
result3 = data.xpath("//ul")
result4 = result3.xpath("./li/a").extract()
# 选取当前节点的父节点，获取父节点的class属性值
result4 = result3.xpath("./li/a")
result5 = result4.xpath("../@class").extract() # ['item-1', 'item-2', 'item-3', 'item-4', 'item-5']
# 获取第三个<li>标签的节点
## 方法1
result6 = data.xpath("//li[3]").extract() # 索引从1开始
## 方法2
result7 = data.xpath("//li")[2].extract()
# 通过定位属性的方法获取第四个<a>标签
result8 = data.xpath("//a[@href='link4.html']").extract() #['<a href="link4.html">Fouth</a>']
# 用属性定位标签，获取第四个<a>标签的文本内容
result9 = data.xpath("//a[@href='link4.html']/text()").extract() # ['Fouth']
# 获取第五个<a>标签的href属性值
result10 = data.xpath("//li[5]/a/@href").extract() # ['link5.html']
result10

```

# Script

```python
script_tags = soup.find_all('script')
with open('scripts_content.txt', 'w', encoding='utf-8') as f:
    # 遍历所有的 <script> 标签
    for i, script_tag in enumerate(script_tags):
        # 提取每个 <script> 标签的文本内容
        script_content = script_tag.text
        # 将内容写入文本文件
        f.write(f"Script {i+1}:\n")
        f.write(script_content)
        f.write("\n\n")
```

```python
html = HTMLParser(response.text)
data = html.css("script[type='text/javascript']")
js_objects = []
for script in data:
    try:
        script_text = script.text()
        if "COMPANY_POSITIONS_DATA" in script_text:
            company_positions_data = chompjs.parse_js_objects(script_text)
            for item in company_positions_data:
                js_objects.append(item)
    except:
        pass
company_positions = js_objects[1]
len(company_positions)
```

```python
data = [{'id': 'BI', 'location': {'name': 'USA'}}, {'id': 'BB', 'location': {'name': 'UK'}}, {'id': 'BC', 'location': {'name': 'REMOTE'}}]

parameters = {"location.name": ["USA", "REMOTE"]}
for key, values in parameters.items():
    path = key.split(".")
    for item in data:
        current_data = item
        print(current_data)
        for i in range(len(path)):
            current_key = path[i]
            print(current_data[current_key]) 
            if current_key in current_data:
                if i == len(path) - 1:
                    if current_data[current_key] in values:
                        print(item['id'])
                else:
                    current_data = current_data[current_key]
                    continue
            break

```

# 动态网站

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# 启动 Chrome 浏览器
options = webdriver.EdgeOptions()
options.use_chromium = True  # 使用 Chromium
options.add_argument('--headless')  # 添加无头模式参数，不弹出浏览器页面
driver = webdriver.Edge(options=options)
driver = webdriver.Chrome()

# 打开网页
url = "https://fa-evmr-saasfaprod1.fa.ocs.oraclecloud.com/hcmUI/CandidateExperience/en/sites/CX_1/requisitions?lastSelectedFacet=LOCATIONS&latitude=42.464161155857674&location=239+School+St%2C+Middlesex%2C+MA%2C+United+States&longitude=-71.42785067776921&mode=geolocation&radius=25&radiusUnit=MI&selectedLocationsFacet=300000000480126"
driver.get(url)

# 等待页面加载完毕
driver.implicitly_wait(3)  # 等待时间可根据需要调整

# 获取页面源代码
html = driver.page_source

# 关闭浏览器
driver.quit()

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(html, 'html.parser')
soup.find("span", class_='job-tile__title')
# 打印网页内容
#print(soup.prettify())
```
